```python
import pandas as pd
custdata="C:\\Users\\VladimirTito\\Desktop\\dataIAA\\event_calendar_2409.csv"
data3=open(custdata,"r",encoding='utf8')
cols=data3.readline().strip().split(",")
print(data3)
n_cols=len(cols)
```

```python
counter=0
main_dict={}
for col in cols:
    main_dict[col]=[]
main_dict
```

```python
xd=0;i=0
for line in data3:
    value=line.encode('utf8').strip().split(",".encode('utf8'))
    #print("\n",line)
    j=0;
    while((j+1)<=len(value)):
        #print(j," ",i//32," ",i%32," ",list(main_dict.keys())[i%32]," ",xd,"  ",value[j])
        if(xd!=0):
            main_dict[cols[i%32]][i//32]=main_dict[cols[i%32]][i//32] +" "+value[j].decode("utf-8", "ignore")
        else:
            main_dict[cols[i%32]].append(value[j].decode("utf-8", "ignore"))
            counter+=1;i+=1
        if ((i%32==15 and xd==0) or (i%32==14 and xd==1)):
            if not ((value[j][:4]=='"<r>'.encode('utf8') and value[j][len(value[j])-5:]=='</r>"')
               or (value[j][:4]=='"<t>'.encode('utf8') and value[j][len(value[j])-5:]=='</t>"')):
                if(value[j][:4]=='"<r>'.encode('utf8')
                   or value[j][:4]=='"<t>'.encode('utf8')):
                    xd=1;i-=1
                elif(value[j][len(value[j])-5:]=='</r>"'.encode('utf8') or value[j][len(value[j])-5:]=='</t>"'.encode('utf8')):
                    xd=0;i+=1
        else:
            if (value[j][:1]=='"'.encode('utf8') and value[j].count('"'.encode('utf8'))==1 and xd==0):
                xd=1;i-=1
            elif (value[j][len(value[j])-1:]=='"'.encode('utf8') and value[j].count('"'.encode('utf8'))==1):
                xd=0;i+=1
        j+=1
print("La data set tiene %d filas y %e columnas" % (counter,n_cols))
```

```python
def sacar_cad(cad1,cad2):
    c=cad1
    while (c.count(cad2)>0):
        gg=c.find(cad2)
        c=c[:gg]+c[gg+1:]
    return c
```

```python
def tipo_event(cad1,cad2):
    tipo=["convocatoria","webinar","conferencia","curso","charla","seminario","congreso","workshop","fullday","programa","tertulia"]
    c1=cad1.lower();c2=cad2.lower()
    for i in tipo:
        if(c1.find(i) != -1 or c2.find(i) != -1):
            return i
        if i=="curso":
            if(c1.find("pde") != -1 or c2.find("pde") != -1 or
              c2.find("sesion 1") != -1 or c2.find("sesiones") != -1 or
              c2.find("sessions") != -1 or c2.find("course") != -1 or
              (c2.find("every") != -1 and c2.find("session") != -1)):
                return i
        elif i=="seminario":
            if(c1.find("seminar") != -1 or c2.find("seminar") != -1):
                return i
        elif i=="charla":
            if(c1.find("talk") != -1 or c2.find("talk") != -1):
                return i
        elif i=="webinar":
            if(c2.find("online") != -1 and (c2.find("sesión") != -1 or
               c2.find("session") != -1)):
                return i
```

```python
tipo=[]
for i in range(len(df.index)):
    tipo.append(tipo_event(df[list(df)[13]][i],df[list(df)[14]][i]))
df["event_tip"]=tipo
```

```python
def exp_event(cad):
    lista=["expositor","presentado","join","presenter"]
    a1=cad.lower();a2=cad;b1=[];b2=[];e=""
    for i in lista:
        if a1.find(i)!=-1:
            d=a1.find(i) + len(i) +1
            b1=a1[d:d+40].strip().split(" ")
            b2=a2[d:d+40].strip().split(" ")
            j=len(b1)    
            while j>0:
                if b1[j-1][:len(b1[j-1])-1]==b2[j-1][:len(b2[j-1])-1]:
                    b2.pop(j-1)
                j-=1
            for i in b2[:len(b2)]:
                e=e+" "+i
    return e
```

```python
event_exp=[]
for i in range(len(df.index)):
    event_exp.append(exp_event(df[list(df)[14]][i]))
df["event_exp"]=event_exp
```
